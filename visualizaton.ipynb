```python
#configures matplotlib to show figures embedded in the notebook
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

### Example

```python
x = np.linspace(-5, 5, 20)
y = x ** 2

figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('parabola function')
show()
```

The figure above is a parabola with vertix (0,0) and open up. The symmetry axis is $x=0$.

```python
subplot(1,2,1)
plot(x, y, 'g--')
xlabel('x')
ylabel('y')

subplot(1,2,2)
plot(y, x, 'r*-')
xlabel('x') 
ylabel('y');
```

The figure above is an example for plotting multiple figures in a specified rows/columns.
